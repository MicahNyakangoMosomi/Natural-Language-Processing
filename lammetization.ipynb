```python
import spacy
```

```python
text = "We were supposed to go to school today but it rained, so we never went"
nlp = spacy.load("en_core_web_md")
```

```python
doc = nlp(text)
print("{:<10} {:<10}".format("text","Lemma"))
print("_"*20)
for token in doc:
    print("{:<10} => {:<10}".format(token.text,token.lemma_))
    
```

```python
filePath = "./data/article.txt"
with open(filePath,"r") as f:
    text = f.read()

print(text)
```

```python
doc2 = nlp(text)
print(doc2)

print("{:<20} {:<20}".format("No:", "Sentence"))
for i,sent in enumerate(doc2.sents,1):
    lemma = [word.lemma_ for word in sent ]
    new_sentence = " ".join(lemma)
    
    new_doc = nlp(new_sentence)
    no_stop_word = [word.text for word in new_doc if not word.is_stop and not word.is_punct]
    new_sentence = " ".join(no_stop_word)
    print(f"{i} => {new_sentence}")
```
