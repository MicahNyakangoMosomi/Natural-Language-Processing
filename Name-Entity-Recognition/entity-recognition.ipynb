```python
import spacy
```

```python
# get the text 
path = "../data/article.txt"
with open(path,"r") as f:
    text = f.read()

print(text)

```

```python
nlp = spacy.load("en_core_web_md")
# nlp is a factory.
# Text goes in → structured linguistic objects come out
docs = nlp(text)
print(docs)
```

```python
for token in docs:
    print(token)
```

```python
for ent in docs.ents:
    print(f"{ent.text} => {ent.label_}")
```

```python
# 
nlp_labels = nlp.get_pipe("ner").labels
print(nlp_labels)
```
