```python
# import spacy 
import spacy 
```

```python
# create a nlp object using spacy.load and correspond it with the en_core_web_sm
nlp = spacy.load("en_core_web_sm")
```

```python
# read the file 
path = "data/article.txt"
with open(path,'r') as f:
    text = f.read()
    print(text)
```

```python
# create a doc object 
doc = nlp(text)
```

```python
print(len(text))
print(len(doc))
```

```python
print("------loop for text ---------")
for token in text[:101]:
    print(token)
```

```python
print("------loop for doc ---------")
for token in doc[:101]:
    print(token)
```

```python
for i,sent in enumerate(doc.sents,1) :
    print(f"{i}: {sent}")
```
